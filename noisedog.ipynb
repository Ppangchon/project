```python
image = cv2.imread("/content/drive/My Drive/noise/noisydog.png")
```

```python
image_bw = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
noiseless_image_bw = cv2.fastNlMeansDenoising(image, None, 20, 7, 21)
```

```python
titles = ['Original Image (grayscale)','Image after removing the noise (grayscale)']
images = [image,image_bw, noiseless_image_bw]
plt.figure(figsize=(13,5))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```
